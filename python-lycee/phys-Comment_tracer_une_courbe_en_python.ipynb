# Comment tracer une courbe en python ?

Pour afficher une courbe, on utilisera la bibliothèque matplotlib, très complète. Vous pouvez aller voir quelques exemples sur la page de la bibliothèque ici.

il faut commencer par importer la bibliothèque matplotlib.pyplot grâce à la commande



```python
import matplotlib.pyplot as plt
```

Il suffit ensuite de disposer de deux listes de même taille par exemple les lignes suivantes permettent de construire une liste de valeurs de x comprises entre 0 et 10  et une liste de valeurs de y telles que $y = 2 x^2 - 4$.


```python
x1 = [ ]
y1 = [ ]
for i in range(11):
    x1.append(i)
    y1.append( 2 * x1[i] * x1[i] - 4)


#ou avec plus de points
x2 = [ ]
y2 = [ ]
for i in range(1001):
    x2.append(i * 10 / 1000)
for val in x2:
    y2.append( 2 * val ** 2 - 4)
```

Les fonctions plot et show permettent ensuite respectivement de construire une courbe et d'afficher le graphique.


```python
%matplotlib inline 
# la ligne précédente est specifique à jupyter à ne pas mettre sur pyzo

plt.plot(x2,y2)
plt.show( )
```


Sur un graphique, on peut ensuite ajouter un certain nombre de choses:

* un texte sur chaque axe


```python
plt.plot(x2,y2)

plt.xlabel("grandeur en abscisses")
plt.ylabel("grandeur en ordonnées")

plt.show( )
```

**Remarque:** plt.xlabel est le nom d'une fonction. Même si la liste mise en abscisses est le temps, cette fonction conserve son nom!

```python
t = [ ]
c = [ ]
for i in range(1001):
    t.append(i)
    c.append( 0.1 * 2**(-t[i]/100))
    
plt.plot(t,c)

plt.xlabel("t")
plt.ylabel("c")

plt.show( )
```

* un titre pour le graphique

```python
plt.plot(x2,y2)

plt.title("Titre du graphique")

plt.show( )
```


* la couleur de la courbe
* le style de la courbe


La ligne suivante permet par exemple de tracer les points sans les relier sous la forme de croix de couleur verte.


```python
plt.plot(x1,y1,"+",color="green")

plt.show()
```

La ligne suivante permet par exemple de tracer les points sous la forme de cercles et de relier les points par une ligne pointillée de couleur rouge.

```python
plt.plot(x1,y1,"o:",color="red")

plt.show()
```
