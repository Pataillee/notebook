# Comment tracer une courbe en python ?

Pour afficher une courbe, on utilisera la bibliothèque matplotlib, très complète. Vous pouvez aller voir quelques exemples sur la page de la bibliothèque ici.

il faut commencer par importer la bibliothèque matplotlib.pyplot grâce à la commande



```python
import matplotlib.pyplot as plt
```

Il suffit ensuite de disposer de deux listes de même taille par exemple les lignes suivantes permettent de construire une liste de valeurs de x comprises entre 0 et 10  et une liste de valeurs de y telles que $y = 2\times 2 - 4$.


```python
x = [ ]
y = [ ]
for i in range(11):
    x.append(i)
    y.append( 2 * x[i] * x[i] - 4)

#ou avec plus de points

x = [ ]
y = [ ]
for i in range(1001):
    x.append(i * 10 / 1000)
for val in x:
    y.append( 2 * val ** 2 - 4)
```

Les fonctions plot et show permettent ensuite respectivement de construire une courbe et d'afficher le graphique.


```python
%matplotlib inline
plt.plot(x,y)
plt.show( )
```


Sur un graphique, on peut ensuite ajouter un certain nombre de choses:

un texte sur chaque axe


```python
plt.xlabel("grandeur en abscisses")
plt.ylabel("grandeur en ordonnées")
```

Un titre pour le graphique

```python
plt.title("Titre du graphique")
```


On peut aussi jouer sur la couleur, et le style de la courbe. La ligne suivante permet par exemple de tracer les points sans les relier sous la forme de croix de couleur rose.



```python
plt.plot(x,y,"+",color="pink")
```


Vous pouvez avoir des exemples et des informations plus complètes sur la documentation en ligne de la bibliothèque matplotlib, sur les fiches méthodes de la section suivante de ce cours ou vous pouvez utiliser la fonction help de python, par exemple help(plt.plot), pour avoir une liste de ce qui est possible.
